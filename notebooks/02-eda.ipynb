# 02 — Análisis Exploratorio (EDA)

Respondemos preguntas de seguridad y de negocio sobre el dataset limpio.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../datasets/firewall_clean.csv')
len(df)
```

## Seguridad — ¿Cuáles son las IP más activas?

```python
top_ips = df['src_ip'].value_counts().head(10)
top_ips.plot(kind='barh', title='Top 10 IPs por nº de eventos')
plt.gca().invert_yaxis(); plt.tight_layout(); plt.show()
top_ips
```

## ¿Qué puertos son los más atacados?

```python
df['dst_port'].value_counts().head(10).plot(kind='bar', title='Puertos destino más frecuentes'); plt.tight_layout(); plt.show()
```

## ¿Qué países generan más eventos?

```python
df['country'].value_counts().head(10).plot(kind='bar', title='Eventos por país'); plt.tight_layout(); plt.show()
```

## ¿Qué horas tienen más actividad?

```python
df['hora'].value_counts().sort_index().plot(kind='line', marker='o', title='Actividad por hora del día'); plt.tight_layout(); plt.show()
```

## Negocio — promedio diario y tendencia mensual

```python
print('Promedio diario de eventos:', round(df.groupby('fecha').size().mean(), 1))
df.groupby('mes').size().plot(kind='bar', title='Eventos por mes'); plt.tight_layout(); plt.show()
```
