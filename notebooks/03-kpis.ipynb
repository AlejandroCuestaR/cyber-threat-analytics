# 03 — KPIs

Indicadores clave para la capa ejecutiva del dashboard.

```python
import pandas as pd

df = pd.read_csv('../datasets/firewall_clean.csv')
logins = pd.read_csv('../datasets/login_attempts.csv')
```

```python
kpis = {
    'Eventos totales': len(df),
    'IPs únicas': df['src_ip'].nunique(),
    'Intentos fallidos (login)': int((logins['success'].astype(str).str.lower()=='false').sum()),
    'Eventos críticos': int((df['severity']=='critica').sum()),
    'Eventos denegados': int((df['action']=='deny').sum()),
    'Top atacante': df['src_ip'].value_counts().idxmax(),
    'Eventos del top atacante': int(df['src_ip'].value_counts().max()),
}
for k, v in kpis.items():
    print(f'{k:<28}: {v}')
```
